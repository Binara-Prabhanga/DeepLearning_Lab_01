```python
import numpy as np
```

```python
exponential_array = np.random.exponential(scale=1.0, size=(4,4))
exponential_array
```

```python
large_exponential_array = np.random.exponential(scale=1.0, size=(100000, 1))
```

```python
import matplotlib.pyplot as plt

plt.hist(large_exponential_array, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.uniform(size=100000), density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(np.random.normal(size=100000), density=True, bins=100, histtype="step", color="red", label="Normal")
plt.axis([0, 10, 0, 1.1])  # Adjust axis to focus on the area of interest
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Load your local dataset using pandas
df = pd.read_csv("/content/Au_nanoparticle_dataset.csv")

filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
filtered_df.head(20)
```

```python
mean_values = filtered_df.mean()
std_values = filtered_df.std()
quartiles = filtered_df.quantile([0.25, 0.5, 0.75])
print("Mean values:\n", mean_values)
print("Standard Deviation:\n", std_values)
print("Quartiles:\n", quartiles)
```

```python
filtered_df.hist(figsize=(15, 6), layout=(1, 4), bins=30)
plt.show()
```

```python
import seaborn as sns

g = sns.pairplot(filtered_df)
plt.show()
```

```python
g = sns.PairGrid(filtered_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot, cmap="Blues_d")
plt.show()
```
